Normal (Gaussian) Distribution

```python
from numpy import random

x = random.normal(size=(2, 3))

print(x)
```

```python
# random.normal()Normal Veri Dağıtımı elde etmek için yöntemi kullanın .

# Üç parametresi vardır:

# loc - (Ortalama) çanın zirvesinin bulunduğu yer.

# scale - (Standart Sapma) grafik dağılımının ne kadar düz olması gerektiği.

# size - Döndürülen dizinin şekli.
```

```python
from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

Visualization of Normal Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```
