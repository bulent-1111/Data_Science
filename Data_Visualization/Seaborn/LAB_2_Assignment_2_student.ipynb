## 3: Plot 20 most dangerous states

```python
kill['state'].value_counts().head(20)
```

```python
state_name = kill['state'].value_counts().head(20).index
state_name
```

```python
state_values = kill['state'].value_counts().head(20).values
state_values
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(state_name, state_values,  data=kill)
ax.set_xticklabels(state_name, rotation=60)
ax.set_title("Most danderous states",color="b",fontsize=15)
ax.set_ylabel("count")
ax.set_xlabel("state")
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+0.1, p.get_height()+0.4));
```

```python
# with seaborn barplot
```

```python
st =  kill['state'].value_counts().head(20)
st
```

```python

st = kill['state'].value_counts().iloc[:20]
fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.countplot(x='state', data=kill, order=st.index)
ax.set_xticklabels(state_name, rotation=60)
ax.set_title("Most danderous states",color="b",fontsize=15)
ax.set_ylabel("count")
ax.set_xlabel("state")
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+0.1, p.get_height()+0.4));
```
