```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1500,1200,1100,1800]
plt.plot(x,y)
plt.show()
```

```python
legend=['januar','February','March','April']
plt.xticks(x,legend)
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.ylabel('Sales in Us Dollars')
plt.title('Montly Sales')
```
