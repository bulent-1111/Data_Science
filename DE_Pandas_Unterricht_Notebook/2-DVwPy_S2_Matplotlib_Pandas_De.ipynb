# Hinzufgen von Text in den Plot
``.text()``

-Hinzufgen von Text in den Plot in Matplotlib

- Die ``matplotlib.pyplot.text()`` - Funktion wird verwendet, um text innerhalb des Grundstck hinzuzufgen . Die syntax zufgt den Text an eine beliebige Position der Achsen hin. Es untersttzt auch mathematische Ausdrcke.

- Syntax: ``matplotlib.pyplot.text(x, y, s, fontdict=None, **kwargs)``

- Parameter:

    - wo x, y  Koordinaten
    - s  text Hinzugefgt werden, innerhalb der Handlung(string)
    - fontdict  optionale parameter. Es berschreibt die Standard-text-Eigenschaften
    - **kwargs  text-Eigenschaften
    
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html

https://www.geeksforgeeks.org/add-text-inside-the-plot-in-matplotlib/

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize=14, ha='center')
    plt.title(f"Subplot: {i}")

plt.suptitle("This is Functional Matplotlib Subplots Created By For Loop")
plt.tight_layout();
```

## Boxplots

- Ein ``Boxplot`` ist eine Mglichkeit, die Verteilung der Daten durch fnf Hauptdimensionen statistisch darzustellen:

    - Minimum: Die kleinste Zahl im Datensatz ohne die Ausreier.

    - Drittes Quartil: Mittlere Zahl zwischen Median und Maximum.

## Kreisdiagramm | Tortendiagramm
-Wir knnen Tortendiagramme in Matplotlib erstellen, indem wir das Schlsselwort `kind=pie` bergeben.

```python
plt.figure(figsize=(10, 8))
y = [35, 25, 25, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, 
        labels=labels, 
        labeldistance=0.7, 
        autopct="%.1f", 
        textprops={'color':"black"});
```

- Das obige Bild ist nicht sehr klar, die Zahlen und der Text berschneiden sich in einigen Fllen. Lassen Sie uns ein paar nderungen vornehmen, um die Optik zu verbessern:

    - Entfernen Sie die Textbeschriftungen auf dem Tortendiagramm, indem Sie die Legende bergeben, und fgen Sie sie mit ``plt.legend()`` als separate Legende hinzu.
    - Verschieben Sie die Prozentstze so, dass sie direkt auerhalb des Tortendiagramms sitzen, indem Sie den Parameter ``pctdistance`` bergeben.
    - bergeben Sie einen benutzerdefinierten ``colors`` fr Kontinente, indem Sie den Farbparameter bergeben.
    - Lsen Sie das Kreisdiagramm auf, um die drei untersten Kontinente (ananas, Apples, Dates and Cherries) hervorzuheben, indem Sie den Explosionsparameter bergeben.

```python
plt.figure(figsize=(10, 8))
y = [35, 25, 25, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]
explode = [0.1, 0, 0, 0]  
plt.pie(y, 
        labels=labels, 
        explode=explode,
        pctdistance=0.66,
        labeldistance=None, 
        autopct="%.1f",
        textprops={'color':"black"})
plt.legend();
```

```python
plt.figure(figsize=(10, 8))
colors_ = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_ = [0, 0, 0.2, 0.1] 

y = [25, 35, 25, 15]
labels = ["Bananas", "Apples", "Cherries", "Dates"]

plt.pie(y, 
        labels=labels,
        labeldistance=0.7, 
        autopct="%.1f", 
        startangle=90, 
        shadow=True, 
        pctdistance=1.1, 
        colors=colors_, 
        explode=explode_,
        textprops={'color':"black"})
plt.legend();
```

# 3D Plotten in Mathplotlib

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(12, 8))
ax = plt.axes(projection='3d')
```

```python
# Daten fr eine dreidimensionale Linie
linie_z = np.linspace(0, 15, 1000)
linie_x = np.sin(linie_z)
linie_y = np.cos(linie_z)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot3D(linie_x, linie_y, linie_z, 'magenta');
```

```python
# Data for three-dimensional scattered points

zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Reds');
```
